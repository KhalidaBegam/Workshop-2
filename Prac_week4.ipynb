```python

# Import necessary libraries
import pandas as pd
import wget
import zipfile
import os
import urllib.request

```

```python
# Define the URL for the dataset
url = 'https://archive.ics.uci.edu/dataset/368/facebook+metrics/dataset_Facebook.csv'

```

```python
# Load the dataset into a pandas DataFrame
df = pd.read_csv('dataset_Facebook.csv', delimiter=';')
df
```

```python
df_subset = df[['Type', 'Lifetime Post Total Reach']]

# Step 2: Calculate the average lifetime post total reach for each type
average_reach = df_subset.groupby('Type').mean().reset_index()

# Step 3: Write the result to a new CSV file
output_file_path = 'average_lifetime_post_total_reach.csv'
average_reach.to_csv(output_file_path, index=False)

print(f"Average Lifetime Post Total Reach calculated and saved to '{output_file_path}'.")
```
